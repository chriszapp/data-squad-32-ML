#### Drop irrelevant rows

Source: https://www.kaggle.com/alphaepsilon/housing-prices-dataset

Based on the documentation of the dataset, I decided to drop the following columns. These are not quantitative columns and will not contribute to our analysis.

```python
house.drop(['PoolQC','LotShape','LandContour','Condition2','MiscFeature','MiscVal','MoSold','YrSold'],axis=1,inplace=True)
```

#### Correlation Matrix with HeatMap

```python
corrmat = house.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(30,20))
g=sns.heatmap(house[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
feat = house.corr()['SalePrice']
feat= feat[feat > 0.3]
cols=list(feat.index)
cols
```

The observation of the correlation matrix allows to observe the features that influentiate our final target the most. From now on, the dataset will only contain the features with a correlation to SalePrice above 0.5. 

```python
house=house[cols]
```

To assure independence between the features, from the pre-selected, I will also remove the ones that have a high correlation between them(<0.8).

```python
house.drop(['1stFlrSF','GarageCars','TotRmsAbvGrd','GarageYrBlt'],axis=1,inplace=True)
```

#### Outliers

To this analysis, I believe that the outliers are important. There might be houses in the future that have unusual features and I want the model to be prepared for that situation.

#### Null Values and Data types

```python
house.isna().sum()
```

```python
(house['LotFrontage']==0).value_counts()
```

The features LotFrontage and MasVnrArea have 259 and 8 null values, respectively. They contain areas measure, so the null values should be replaced by 0.

```python
house.fillna({'LotFrontage':0,'MasVnrArea':0},inplace=True)
```

All the features selected are numerical and there are no null values.
